KNN CLASSIFICATION ALGORITHM

```python
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns
df=sns.load_dataset('iris')
df
```

```python
df.isnull().sum()
```

```python
X=df.iloc[:,0:4]
y=df.iloc[:,4:5]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN_classifier=KNeighborsClassifier()
KNN_classifier.fit(X_train,y_train)
predictions=KNN_classifier.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

#performance metrci check
confusionMetric=confusion_matrix(y_test,predictions)
print(confusionMetric)
```

```python
#calculating the accuracy
accuracy_score=accuracy_score(y_test,predictions)
print(accuracy_score)
```

```python
#calculating error for K values between 1 and 20
for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    pred_i=pd.DataFrame(pred_i.shape)
    misclassified = (pred_i != y_test).sum()
    misclassification_rate = misclassified / len(y_test)
    misclassification_rates.append(misclassification_rate)
    print(pd.DataFrame(pred_i.shape))
    print(y_test)


print(misclassification_rates)
```
