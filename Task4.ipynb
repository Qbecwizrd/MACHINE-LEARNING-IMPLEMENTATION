```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('toyota.csv')
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=1)
```

```python
df
```

```python
plt.scatter(x=df['Age'],y=df['Price'],c="red")
plt.title="scaltter plot of price vs age"
plt.show()
```
